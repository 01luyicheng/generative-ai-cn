### 控制思考预算

您可以在 `ThinkingConfig` 中设置可选的 `thinking_budget` 参数，以控制和配置模型针对给定用户提示的思考量。`thinking_budget` 设置了用于某些任务推理的 Token 数量上限，允许用户在响应质量和速度之间进行权衡。

**注意事项**

- 默认情况下，模型会自动控制思考量，最多使用 8192 个 Token。
- 您可以设置的最大思考预算为 `32768` 个 Token，最小为 `128` 个 Token。

然后使用 `generate_content` 或 `generate_content_stream` 方法，通过 `thinking_config` 发送生成内容的请求。

````
THINKING_BUDGET = 1024  # @param {type: "integer"}

response = client.models.generate_content(
    model=MODEL_ID,
    contents="How many R's are in the word strawberry?",
    config=GenerateContentConfig(
        thinking_config=ThinkingConfig(
            thinking_budget=THINKING_BUDGET,
        )
    ),
)

display(Markdown(response.text))
````

您也可以选择打印模型响应中的 usage_metadata 和 Token 计数。

````
print(f"prompt_token_count: {response.usage_metadata.prompt_token_count}")
print(f"candidates_token_count: {response.usage_metadata.candidates_token_count}")
print(f"thoughts_token_count: {response.usage_metadata.thoughts_token_count}")
print(f"total_token_count: {response.usage_metadata.total_token_count}")
````

### 查看摘要思考内容

您可以选择设置 `include_thoughts` 标志，使模型除最终答案外，还会生成并返回其"思考过程"摘要。

在本示例中，您使用 `generate_content` 方法发送带有摘要思考的内容生成请求。模型会响应多个部分，包括思考过程和模型响应。您可以通过检查 `part.thought` 字段来判断某个部分是否为思考内容。

````
response = client.models.generate_content(
    model=MODEL_ID,
    contents="How many R's are in the word strawberry?",
    config=GenerateContentConfig(
        thinking_config=ThinkingConfig(
            include_thoughts=True,
        )
    ),
)

for part in response.candidates[0].content.parts:
    if part.thought:
        display(
            Markdown(
                f"""## Thoughts:
         {part.text}
        """
            )
        )
    else:
        display(
            Markdown(
                f"""## Answer:
         {part.text}
        """
            )
        )
````

### 生成内容流

默认情况下，模型在完成整个生成过程后才返回响应。您也可以使用 `generate_content_stream` 方法在生成过程中流式传输响应，模型会在生成后立即返回响应的各个数据块。

本示例展示了如何在 `generate_content_stream` 方法中设置 `include_thoughts` 和 `thinking_budget`。

````
THINKING_BUDGET = 1024  # @param {type: "integer"}
INCLUDE_THOUGHTS = True  # @param {type: "boolean"}

prompt = """
A bat and a ball cost $1.10 in total.
The bat costs $1.00 more than the ball.
How much does the ball cost?
"""

thoughts = ""
answer = ""

for chunk in client.models.generate_content_stream(
    model=MODEL_ID,
    contents=prompt,
    config=GenerateContentConfig(
        thinking_config=ThinkingConfig(
            thinking_budget=THINKING_BUDGET,
            include_thoughts=INCLUDE_THOUGHTS,
        )
    ),
):

    for part in chunk.candidates[0].content.parts:
        if not part.text:
            continue
        elif part.thought:
            if not thoughts:
                display(Markdown("## Thoughts"))
            display(Markdown(part.text))
            thoughts += part.text
        else:
            if not answer:
                display(Markdown("## Answer"))
            display(Markdown(part.text))
            answer += part.text
````

## 配置模型参数

您可以在每次向模型发送的请求中包含参数值，以控制模型生成响应的方式。对于不同的参数值，模型可能会生成不同的结果。您可以尝试不同的模型参数，观察结果如何变化。

- 了解更多关于[实验参数值](https://cloud.google.com/vertex-ai/generative-ai/docs/learn/prompts/adjust-parameter-values)的信息。

- 查看所有 [Gemini API 参数](https://cloud.google.com/vertex-ai/generative-ai/docs/model-reference/inference#parameters)的列表。


````
response = client.models.generate_content(
    model=MODEL_ID,
    contents="Tell me how the internet works, but pretend I'm a puppy who only understands squeaky toys.",
    config=GenerateContentConfig(
        temperature=2.0,
        top_p=0.95,
        candidate_count=1,
        max_output_tokens=8000,
        thinking_config=thinking_config,
    ),
)

display(Markdown(response.text))
````

## 设置系统指令

[系统指令](https://cloud.google.com/vertex-ai/generative-ai/docs/learn/prompts/system-instruction-introduction)允许您引导模型的行为。通过设置系统指令，您可以为模型提供额外的上下文，使其更好地理解任务、提供更具针对性的响应，并在整个用户交互过程中遵循相关准则。

````
system_instruction = """
  You are a helpful language translator.
  Your mission is to translate text in English to Spanish.
"""

prompt = """
  User input: I like bagels.
  Answer:
"""

response = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt,
    config=GenerateContentConfig(
        system_instruction=system_instruction,
        thinking_config=thinking_config,
    ),
)

display(Markdown(response.text))
````

## 安全过滤器

Gemini API 提供了安全过滤器，您可以在多个过滤类别中进行调整，以限制或允许特定类型的内容。您可以使用这些过滤器来调整适合您使用场景的内容标准。详情请参阅[配置安全过滤器](https://cloud.google.com/vertex-ai/generative-ai/docs/multimodal/configure-safety-filters)页面。

当您向 Gemini 发送请求时，内容会被分析并被赋予安全评级。您可以通过打印模型响应来检查生成内容的安全评级。

安全设置默认为 `OFF`，默认屏蔽阈值为 `BLOCK_NONE`。

有关安全过滤器的更多示例，请参阅[此笔记本](https://github.com/GoogleCloudPlatform/generative-ai/blob/main/gemini/responsible-ai/gemini_safety_ratings.ipynb)。

您可以使用 `safety_settings` 为每次向 API 发送的请求调整安全设置。以下示例演示如何将所有类别的屏蔽阈值设置为 `BLOCK_LOW_AND_ABOVE`：

````
system_instruction = "Be as mean and hateful as possible."

prompt = """
    Write a list of 5 disrespectful things that I might say to the universe after stubbing my toe in the dark.
"""

safety_settings = [
    SafetySetting(
        category=HarmCategory.HARM_CATEGORY_DANGEROUS_CONTENT,
        threshold=HarmBlockThreshold.BLOCK_LOW_AND_ABOVE,
    ),
    SafetySetting(
        category=HarmCategory.HARM_CATEGORY_HARASSMENT,
        threshold=HarmBlockThreshold.BLOCK_LOW_AND_ABOVE,
    ),
    SafetySetting(
        category=HarmCategory.HARM_CATEGORY_HATE_SPEECH,
        threshold=HarmBlockThreshold.BLOCK_LOW_AND_ABOVE,
    ),
    SafetySetting(
        category=HarmCategory.HARM_CATEGORY_SEXUALLY_EXPLICIT,
        threshold=HarmBlockThreshold.BLOCK_LOW_AND_ABOVE,
    ),
]

response = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt,
    config=GenerateContentConfig(
        system_instruction=system_instruction,
        safety_settings=safety_settings,
        thinking_config=thinking_config,
    ),
)

# Response will be `None` if it is blocked.
print(response.text)
# Finish Reason will be `SAFETY` if it is blocked.
print(response.candidates[0].finish_reason)
# Safety Ratings show the levels for each filter.
for safety_rating in response.candidates[0].safety_ratings:
    print(safety_rating)
````

## 控制生成输出

[受控生成](https://cloud.google.com/vertex-ai/generative-ai/docs/multimodal/control-generated-output)允许您定义响应模式，以指定模型输出的结构、字段名称以及每个字段的预期数据类型。

响应模式在 `config` 的 `response_schema` 参数中指定，模型输出将严格遵循该模式。

您可以以 [Pydantic](https://docs.pydantic.dev/) 模型或 [JSON](https://www.json.org/json-en.html) 字符串的形式提供模式，模型将根据 `response_mime_type` 的设置，以 JSON 或 [Enum](https://docs.python.org/3/library/enum.html) 格式进行响应。

有关受控生成的更多示例，请参阅[此笔记本](https://github.com/GoogleCloudPlatform/generative-ai/blob/main/gemini/controlled-generation/intro_controlled_generation.ipynb)。

````
from pydantic import BaseModel


class Recipe(BaseModel):
    name: str
    description: str
    ingredients: list[str]


response = client.models.generate_content(
    model=MODEL_ID,
    contents="List a few popular cookie recipes and their ingredients.",
    config=GenerateContentConfig(
        response_mime_type="application/json",
        response_schema=Recipe,
        thinking_config=thinking_config,
    ),
)

print(response.text)
````

您可以将响应字符串解析为 JSON，或使用 `parsed` 字段以对象或字典形式获取响应。

````
parsed_response: Recipe = response.parsed
print(parsed_response)
````

您也可以在 Python 字典中定义响应模式。您只能使用以下支持的字段，其他所有字段均会被忽略。

- `enum`
- `items`
- `maxItems`
- `nullable`
- `properties`
- `required`

在本示例中，您指示模型分析产品评论数据、提取关键实体、执行情感分类（多选）、提供额外说明，并以 JSON 格式输出结果。


````
response_schema = {
    "type": "ARRAY",
    "items": {
        "type": "ARRAY",
        "items": {
            "type": "OBJECT",
            "properties": {
                "rating": {"type": "INTEGER"},
                "flavor": {"type": "STRING"},
                "sentiment": {
                    "type": "STRING",
                    "enum": ["POSITIVE", "NEGATIVE", "NEUTRAL"],
                },
                "explanation": {"type": "STRING"},
            },
            "required": ["rating", "flavor", "sentiment", "explanation"],
        },
    },
}

prompt = """
  Analyze the following product reviews, output the sentiment classification, and give an explanation.

  - "Absolutely loved it! Best ice cream I've ever had." Rating: 4, Flavor: Strawberry Cheesecake
  - "Quite good, but a bit too sweet for my taste." Rating: 1, Flavor: Mango Tango
"""

response = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt,
    config=GenerateContentConfig(
        response_mime_type="application/json",
        response_schema=response_schema,
        thinking_config=thinking_config,
    ),
)

print(response.text)
````

## 搜索即工具（基础）

[基础](https://cloud.google.com/vertex-ai/generative-ai/docs/multimodal/ground-gemini)功能让您能够将现实世界的数据与 Gemini 模型相连接。

通过将模型响应与 Google 搜索结果相关联，模型可以在运行时访问超出其训练数据范围的信息，从而提供更准确、更及时、更相关的响应。

使用基于 Google 搜索的基础功能，您可以提高模型响应的准确性和时效性。从 Gemini 2.0 开始，Google 搜索可作为工具使用，这意味着模型可以自行决定何时使用 Google 搜索。

有关基础功能的更多示例，请参阅[此笔记本](https://github.com/GoogleCloudPlatform/generative-ai/blob/main/gemini/grounding/intro-grounding-gemini.ipynb)。

### Google 搜索

您可以添加 `tools` 关键字参数，其中包含带有 `GoogleSearch` 的 `Tool`，以指示 Gemini 先使用提示执行 Google 搜索，再根据网络搜索结果构建答案。

````
google_search_tool = Tool(google_search=GoogleSearch())

response = client.models.generate_content(
    model=MODEL_ID,
    contents="What is the current temperature in Austin, TX?",
    config=GenerateContentConfig(
        tools=[google_search_tool],
        thinking_config=thinking_config,
    ),
)

display(Markdown(response.text))

print(response.candidates[0].grounding_metadata)

HTML(response.candidates[0].grounding_metadata.search_entry_point.rendered_content)
````

## 函数调用

Gemini 中的[函数调用](https://cloud.google.com/vertex-ai/docs/generative-ai/multimodal/function-calling)允许开发者在代码中创建函数描述，然后将该描述随请求一起传递给语言模型。

您可以提交 Python 函数进行自动函数调用，模型将运行该函数并以 Gemini 生成的自然语言返回输出。

您也可以提交 [OpenAPI 规范](https://www.openapis.org/)，模型将响应与描述匹配的函数名称以及调用所需的参数。

有关使用 Gemini 进行函数调用的更多示例，请查看此笔记本：[Gemini 函数调用简介](https://github.com/GoogleCloudPlatform/generative-ai/blob/main/gemini/function-calling/intro_function_calling.ipynb)

### Python 函数（自动函数调用）

````
def get_current_weather(location: str) -> str:
    """Example method. Returns the current weather.

    Args:
        location: The city and state, e.g. San Francisco, CA
    """
    weather_map: dict[str, str] = {
        "Boston, MA": "snowing",
        "San Francisco, CA": "foggy",
        "Seattle, WA": "raining",
        "Austin, TX": "hot",
        "Chicago, IL": "windy",
    }
    return weather_map.get(location, "unknown")


response = client.models.generate_content(
    model=MODEL_ID,
    contents="What is the weather like in San Francisco?",
    config=GenerateContentConfig(
        tools=[get_current_weather],
        temperature=0,
        thinking_config=thinking_config,
    ),
)

display(Markdown(response.text))
````

### OpenAPI 规范（手动函数调用）

````
get_destination = FunctionDeclaration(
    name="get_destination",
    description="Get the destination that the user wants to go to",
    parameters={
        "type": "OBJECT",
        "properties": {
            "destination": {
                "type": "STRING",
                "description": "Destination that the user wants to go to",
            },
        },
    },
)

destination_tool = Tool(
    function_declarations=[get_destination],
)

response = client.models.generate_content(
    model=MODEL_ID,
    contents="I'd like to travel to Paris.",
    config=GenerateContentConfig(
        tools=[destination_tool],
        temperature=0,
        thinking_config=thinking_config,
    ),
)

print(response.function_calls[0])
````

## 代码执行

Gemini API 的[代码执行](https://ai.google.dev/gemini-api/docs/code-execution?lang=python)功能使模型能够生成并运行 Python 代码，并从结果中迭代学习，直至得出最终输出。您可以使用此代码执行能力来构建受益于基于代码推理并产生文本输出的应用程序。例如，您可以在解方程或处理文本的应用程序中使用代码执行。

Gemini API 将代码执行作为工具提供，类似于函数调用。
在您将代码执行添加为工具后，模型将自行决定何时使用它。

有关代码执行的更多示例，请参阅[此笔记本](https://github.com/GoogleCloudPlatform/generative-ai/blob/main/gemini/code-execution/intro_code_execution.ipynb)。

````
code_execution_tool = Tool(code_execution=ToolCodeExecution())

response = client.models.generate_content(
    model=MODEL_ID,
    contents="Calculate 20th fibonacci number. Then find the nearest palindrome to it.",
    config=GenerateContentConfig(
        tools=[code_execution_tool],
        temperature=0,
        thinking_config=thinking_config,
    ),
)

display(
    Markdown(
        f"""
## Code

```py
{response.executable_code}
```

### Output

```
{response.code_execution_result}
```
"""
    )
)
````
